```python
import joblib
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
# Load the saved preprocessor and models
preprocessor = joblib.load('preprocessor.pkl')
best_knn_model = joblib.load('best_knn_model.pkl')
best_tree_model = joblib.load('best_tree_model.pkl')
```

```python
df=pd.read_csv(r"C:\Users\chira\.cache\kagglehub\datasets\uciml\pima-indians-diabetes-database\versions\1\diabetes.csv")
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_category'] = df['BMI'].apply(categorize_bmi)
df.head()

 
```

```python
# Test sample input (5 samples from validation set)
val_samples = df.iloc[:5, :].drop(columns=['Outcome'])  # Assuming the first 5 samples for inference

# Predict with KNN
knn_predictions = best_knn_model.predict(val_samples)
print("KNN Predictions:", knn_predictions)

# Predict with Decision Tree
dt_predictions = best_tree_model.predict(val_samples)
print("Decision Tree Predictions:", dt_predictions)

# Print the predicted classes
for i, prediction in enumerate(knn_predictions):
    print(f"Sample {i+1} - KNN Prediction: {prediction}")

for i, prediction in enumerate(dt_predictions):
    print(f"Sample {i+1} - Decision Tree Prediction: {prediction}")
```
